```python
import pandas as pd

# Create dataset
data = {
    "Order ID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    "Customer": ["Alice", "Bob", "Charlie", "Alice", "David", "Charlie", "Bob", "Alice"],
    "Product": ["Laptop", "Phone", "Chair", "Table", "Laptop", "Phone", "Chair", "Phone"],
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Electronics", "Electronics", "Furniture", "Electronics"],
    "Quantity": [2, 1, 4, 1, 1, 2, 2, 1],
    "Price": [800, 500, 100, 250, 800, 500, 100, 500],
    "Total Sales": [1600, 500, 400, 250, 800, 1000, 200, 500]
}

df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
# 1. Create a DataFrame:
#  Load the above dataset into a Pandas DataFrame.
```

```python
# Add a New Column:
# Add a column Discount that gives a 10% discount on total sales.
```

```python
df["Discount"]=df["Total Sales"]*0.10
```

```python
df
```

```python
# Add a New Row:
# Add a new sales record: Order ID = 1009, Customer =
# "Eve", Product = "Tablet", Category = "Electronics",
# Quantity = 1, Price = 300, Total Sales = 300.
```

```python
new_row = {
    "Order ID": 1009,
    "Customer": "Eve",
    "Product": "Tablet",
    "Category": "Electronics",
    "Quantity": 1,
    "Price": 300,
    "Total Sales": 300
}

df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

print(df)
```

```python
# Sorting:
# Sort the DataFrame based on Total Sales in descending order.
```

```python
df[["Customer","Total Sales"]].sort_values(by="Total Sales", ascending=False)
```

```python
# Apply Function:
# Apply a lambda function to create a new column Sales
# Category, where values are labeled as "High" if Total Sales >
# 1000, otherwise "Low".
```

```python
# Create Sales Category column
df["Sales Category"] = df["Total Sales"].apply(lambda x: "High" if x > 1000 else "Low")

print(df)
```
